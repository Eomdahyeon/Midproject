```python
import pandas as pd

# 파일 경로 설정
file_sugup = 'data/first_21_rows.csv'
file_total = 'data/시군구별_연도별_고령인구비율_시각화 사용.csv'

# 파일 불러오기
df_sugup = pd.read_csv(file_sugup, encoding='utf-8')
df_total = pd.read_csv(file_total, encoding='utf-8')

#연도별 총계(국민연금 수급자수) 전처리
df_sugup['총계'] = df_sugup['총계'].str.replace(',', '').astype(int)

# 연도별 65세 이상 인구 총합 계산
df_total['연도'] = df_total['연도'].str.replace(' 년', '').astype(int)
df_total_grouped = df_total.groupby('연도')['65세이상'].sum().reset_index()
df_total_grouped.columns = ['연도', '65세이상총합']

#연도 기준으로 데이터 병합
df_merged = pd.merge(df_sugup[['연도', '총계']], df_total_grouped, on='연도')

# 국민연금 수급률 계산
df_merged['국민연금수급률(%)'] = df_merged['총계'] / df_merged['65세이상총합'] * 100

# 결과 출력
print(df_merged[['연도', '총계', '65세이상총합', '국민연금수급률(%)']])
```

```python
# 파일 저장 경로 (예: '/mnt/data/전처리된_연금데이터.csv')
output_file_path = 'data/고령층 국민연금 수급률_전처리.csv'

# 전처리된 데이터프레임(df_multi_header)을 CSV로 저장
df_merged.to_csv(output_file_path, index=False, encoding='utf-8-sig')

print(f"전처리된 데이터가 {output_file_path}에 저장되었습니다!")
```

```python
import pandas as pd

# 파일 경로
file_total_amount = 'data/df_filtered.csv'     # 고령층 총수급액
file_sugup_count = 'data/df_filtered1.csv'     # 고령층 수급자 수

# 데이터 불러오기
df_total_amount = pd.read_csv(file_total_amount, encoding='utf-8')
df_sugup_count = pd.read_csv(file_sugup_count, encoding='utf-8')

# ',' 제거 및 정수형 변환
df_total_amount['연금수급자 노령'] = df_total_amount['연금수급자 노령'].str.replace(',', '').astype(int)
df_sugup_count['연금수급자 노령'] = df_sugup_count['연금수급자 노령'].str.replace(',', '').astype(int)

# 컬럼명 정리
df_total_amount = df_total_amount.rename(columns={'연금수급자 노령': '고령층총수급액'})
df_sugup_count = df_sugup_count.rename(columns={'연금수급자 노령': '고령층수급자수'})

# 연도 컬럼 정수형으로 변환 (소수점 버림)
df_total_amount['연도'] = df_total_amount['연도'].astype(int)
df_sugup_count['연도'] = df_sugup_count['연도'].astype(int)

# 연도 기준 병합
df_merged = pd.merge(df_total_amount, df_sugup_count, on='연도')

# 평균 수령액 계산 (그대로 float 유지하거나 필요 시 정수로 변환)
df_merged['평균수령액'] = df_merged['고령층총수급액'] / df_merged['고령층수급자수']

# 결과 출력
print(df_merged[['연도', '고령층총수급액', '고령층수급자수', '평균수령액']])
```

```python
output_file = 'data/고령층 국민연금 평균 수령액_전처리.csv'
df_merged.to_csv(output_file, index=False, encoding='utf-8-sig')
print(f"평균수령액 계산 결과가 {output_file}에 저장되었습니다!")
```
