```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 그래프를 그리기위한 기본 설정
# 폰트설정 - 윈도우
plt.rcParams['font.family'] = 'Malgun Gothic'
# 폰트설정 - 맥
# plt.rcParams['font.family'] = 'AppleGothic'
# 그래프의 크기
plt.rcParams['figure.figsize'] = 12, 6
# 글자크기
plt.rcParams['font.size'] = 14
# 적용한 폰트에 - 기호가 없다면 깨진다. - 기호는 설정한 폰트의
# 영향을 받지 않도록 설정한다.
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('data/한국 고령층 빈곤율_전처리.csv')
df
```

```python
# 'Korea' 행 선택 및 '%' 제거 후 float 타입으로 변환
# '빈곤율' 컬럼에 'Korea'인 행의 나머지 연도 컬럼을 선택
korea_row = df[df['빈곤율'] == 'Korea'].iloc[0, 1:]  # 첫 번째 열(빈곤율) 제외하고 나머지 열 선택
korea_row = korea_row.apply(lambda x: float(str(x).replace('%', '')))

# 연도 추출 및 데이터프레임 생성
years = [int(col.replace('년', '')) for col in korea_row.index]
plot_df = pd.DataFrame({'연도': years, '노인빈곤율(%)': korea_row.values})

# 시각화 스타일 설정
sns.set(style='whitegrid', font='Malgun Gothic', font_scale=1.2)

# 라인차트 그리기
plt.figure(figsize=(10, 6))
sns.lineplot(x='연도', y='노인빈곤율(%)', data=plot_df, marker='o', linewidth=2, color='b')

# 제목 및 축 레이블 추가
plt.title('한국 연도별 노인빈곤율 추이')
plt.xlabel('연도')
plt.ylabel('노인빈곤율 (%)')
plt.ylim(38, 50)
plt.grid(True)

# 데이터 포인트 값 표시
for x, y in zip(plot_df['연도'], plot_df['노인빈곤율(%)']):
    plt.text(x, y + 0.5, f"{y}%", ha='center', va='bottom', fontsize=10)

# 그래프 출력
plt.show()
```
