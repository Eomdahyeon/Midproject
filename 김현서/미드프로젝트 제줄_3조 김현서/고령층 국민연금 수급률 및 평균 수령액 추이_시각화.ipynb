```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 그래프를 그리기위한 기본 설정
# 폰트설정 - 윈도우
plt.rcParams['font.family'] = 'Malgun Gothic'
# 폰트설정 - 맥
# plt.rcParams['font.family'] = 'AppleGothic'
# 그래프의 크기
plt.rcParams['figure.figsize'] = 12, 6
# 글자크기
plt.rcParams['font.size'] = 14
# 적용한 폰트에 - 기호가 없다면 깨진다. - 기호는 설정한 폰트의
# 영향을 받지 않도록 설정한다.
plt.rcParams['axes.unicode_minus'] = False
```

```python
# CSV 파일 불러오기
file_rate = 'data/고령층 국민연금 수급률_전처리.csv'
df_rate = pd.read_csv(file_rate, encoding='utf-8')

# 시각화
plt.figure(figsize=(10,6))
plt.plot(df_rate['연도'], df_rate['국민연금수급률(%)'], marker='o', linestyle='-', color='blue', label='국민연금 수급률(%)')
plt.xlabel('연도')
plt.ylabel('수급률 (%)')
plt.title('연도별 국민연금 수급률 추이')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# CSV 파일 불러오기
file_avg = 'data/고령층 국민연금 평균 수령액_전처리.csv'
df_avg = pd.read_csv(file_avg, encoding='utf-8')

# 연도 컬럼 전처리: 소수점 제거, 정수 변환
df_avg['연도'] = df_avg['연도'].astype(str).str.extract(r'(\d{4})').astype(int)

# 연도에서 2025 제외
df_avg = df_avg[df_avg['연도'] != 2025]

# 시각화
plt.figure(figsize=(10,6))
plt.plot(df_avg['연도'], df_avg['평균수령액'], marker='o', linestyle='-', color='green', label='평균수령액')
plt.xlabel('연도')
plt.ylabel('평균수령액')
plt.title('연도별 국민연금 평균수령액 추이')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# CSV 파일 불러오기
file_avg = 'data/고령층 국민연금 평균 수령액_전처리.csv'
df_avg = pd.read_csv(file_avg, encoding='utf-8')

# 연도 전처리 (정수형 변환)
df_avg['연도'] = df_avg['연도'].astype(float).astype(int)

# 연도에서 2025 제외
df_avg = df_avg[df_avg['연도'] != 2025]

# 시각화
plt.figure(figsize=(10,6))
plt.plot(df_avg['연도'], df_avg['평균수령액'], marker='o', linestyle='-', color='green', label='평균수령액')
plt.xlabel('연도')
plt.ylabel('평균수령액')
plt.title('연도별 국민연금 평균수령액 추이')
plt.grid(True)
plt.legend()

# x축 눈금을 연도 리스트로 지정하고, 각 눈금을 45도 회전
plt.xticks(df_avg['연도'].unique(), rotation=45)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 불러오기
file_path = 'data/국민연금 연도별 급여지급액 및 수급자 현황_전처리.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
# 연도, 금액, 수급자 컬럼 확인 및 설정 (예: 'Year', 'Amount', 'Beneficiaries')
# 실제 컬럼명을 확인해 아래 코드에서 수정 필요
year_col = '연도'
amount_col = '고령층총수급액'
beneficiaries_col = '고령층수급자수'

# 시각화
fig, ax1 = plt.subplots(figsize=(10, 6))

# 급여 지급액 (백만원) - 막대그래프
ax1.bar(df[year_col], df[amount_col], color='#c2c2c2', label='급여지급액 (백만원)')
ax1.set_xlabel('연도')
ax1.set_ylabel('급여지급액 (백만원)', color='#c2c2c2')
ax1.tick_params(axis='y', labelcolor='#787878')

# 수급자 수 (천명) - 꺾은선그래프
ax2 = ax1.twinx()
ax2.plot(df[year_col], df[beneficiaries_col], color='#3452f0', marker='o', label='수급자 수')
ax2.set_ylabel('수급자 수', color='#3452f0')
ax2.tick_params(axis='y', labelcolor='#3452f0')

# 제목 및 범례
plt.title('국민연금 연도별 급여지급액 및 수급자 현황')
fig.tight_layout()
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
```

```python
# 컬럼명 설정 (파일의 실제 컬럼명으로 수정해 주세요)
year_col = '연도'
avg_amount_col = '평균수령액'

# '평균수령액' 컬럼 숫자로 변환 (쉼표 등 제거 후 float 변환)
df[avg_amount_col] = df[avg_amount_col].replace({',': ''}, regex=True).astype(float)

# 시각화
plt.figure(figsize=(10, 6))
plt.plot(df[year_col], df[avg_amount_col], color='mediumpurple', marker='o', linestyle='-', label='평균수령액')

# 그래프 꾸미기
plt.xlabel('연도')
plt.ylabel('평균수령액 (원)')
plt.title('국민연금 연도별 평균수령액 추이')
plt.grid(True)
plt.legend()
plt.tight_layout()

# 그래프 출력
plt.show()
```
