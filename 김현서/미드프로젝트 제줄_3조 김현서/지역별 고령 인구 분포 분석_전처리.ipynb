```python
# 시군구별로 전체 인구수와 65세 이상 인구수 합계 계산
# 예를 들어 '행정구역(시군구)별' 컬럼과 '연령별', '2024 년'이 있다고 가정
# 먼저 65세 이상만 필터링
elderly_df = df1[df1['연령별'].isin([f'{i}세' for i in range(65, 101)])]

# 시군구별 65세 이상 인구 합계
elderly_sum = elderly_df.groupby('행정구역(시군구)별')['2024 년'].sum().reset_index(name='65세이상')

# 시군구별 전체 인구 합계
total_sum = df1.groupby('행정구역(시군구)별')['2024 년'].sum().reset_index(name='전체인구')

# 두 데이터프레임 병합
merged_df = pd.merge(total_sum, elderly_sum, on='행정구역(시군구)별', how='left')

# NaN은 0으로 처리 (65세 이상 인구가 없는 경우)
merged_df['65세이상'] = merged_df['65세이상'].fillna(0)

# 고령 인구 비율 계산
merged_df['고령인구비율(%)'] = (merged_df['65세이상'] / merged_df['전체인구']) * 100

# 결과 확인
print(merged_df)

# 결과 저장 (optional)
merged_df.to_csv('data/시군구별_고령인구비율.csv', index=False)
# print("시군구별 고령인구 비율 계산 완료!")
```

```python
# 사용할 연도 컬럼 추출 (예: '2023 년', '2024 년', '2025.04 월' 등)
year_columns = [col for col in df.columns if any(year in col for year in ['2014 년','2015 년','2016 년','2017 년','2018 년','2019 년','2020 년','2021 년','2022 년','2023 년', '2024 년'])]

# 결과 저장용 리스트
results = []

for year_col in year_columns:
    # 65세 이상 필터링
    elderly_df = df[df['연령별'].isin([f'{i}세' for i in range(65, 101)])]

    # 시군구별 65세 이상 인구 합계
    elderly_sum = elderly_df.groupby('행정구역(시군구)별')[year_col].sum().reset_index(name='65세이상')

    # 시군구별 전체 인구 합계
    total_sum = df.groupby('행정구역(시군구)별')[year_col].sum().reset_index(name='전체인구')

    # 병합
    merged_df = pd.merge(total_sum, elderly_sum, on='행정구역(시군구)별', how='left')
    merged_df['65세이상'] = merged_df['65세이상'].fillna(0)

    # 고령 인구 비율 계산
    merged_df['고령인구비율(%)'] = (merged_df['65세이상'] / merged_df['전체인구']) * 100
    merged_df['연도'] = year_col

    results.append(merged_df)

# 연도별 데이터 합치기
final_df = pd.concat(results, ignore_index=True)

# 결과 확인
print(final_df.head())

# 필요 시 CSV로 저장
# final_df.to_csv('data/시군구별_연도별_고령인구비율.csv', index=False)
# print("시군구별 연도별 고령 인구 비율 계산 완료!")
```
