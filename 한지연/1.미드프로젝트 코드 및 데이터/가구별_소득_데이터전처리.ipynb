```python
workingAge_population = pd.read_csv('data/주요_연령계층별_추계인구_생산연령인구__고령인구_전국.csv', encoding='EUC_KR')
workingAge_population
```

```python
workingAge_population['인구종류별'].unique()
```

```python
population_list = ['생산연령인구(천명): 계(15~64세)', '고령인구(천명): 65세+']

workingAge_population.query('인구종류별 in @population_list')
```

```python
# workingAge_population['고령인구비율(%)'] = (workingAge_population['노년층_인구'].fillna(0) / workingAge_population['전국_총인구'].replace(0, np.nan)) * 100
pivot_df = workingAge_population.pivot(
    index='시점',
    columns='인구종류별',
    values='데이터'
).reset_index()

pivot_df.columns.name = None  # '인구종류별' 제거
pivot_df = pivot_df.rename(columns={
    '생산연령인구(천명): 계(15~64세)': '생산연령인구',
    '고령인구(천명): 65세+': '고령인구'
})

clean_df = pivot_df[['시점', '생산연령인구', '고령인구']].copy()
clean_df
```

```python
# fillna(0) → 결측치를 0으로 처리
# replace(0, np.nan) → 분모가 0일 경우는 NaN으로 처리해서 0으로 나누는 오류 방지
clean_df['고령인구비율(%)'] = round((clean_df['고령인구'].fillna(0) / clean_df['생산연령인구'].replace(0, np.nan)) * 100, 1)
clean_df
```

```python
clean_df.to_csv('생산가능인구감소관련_데이터.csv', encoding='utf-8-sig', index=False)
```
