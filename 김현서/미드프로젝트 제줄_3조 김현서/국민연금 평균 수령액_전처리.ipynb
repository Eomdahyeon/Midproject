```python
import pandas as pd
```

```python
df = pd.read_csv('data/2025년_1월_시군구별_1인당_월지급액_평균.csv')
df
```

```python
# 파일 경로 지정
file_path = 'data/2025년_1월_시군구별_1인당_월지급액_평균.csv'

# CSV 파일 로드 (header=[0,1]으로 다중 컬럼 처리)
df1 = pd.read_csv(file_path, encoding='utf-8', header=[0,1])

# 다중 컬럼을 단일 컬럼으로 합치기 (ex: '연금수급자(B) (명)_계')
df1.columns = ['_'.join([str(i) for i in col if i != '']).strip() for col in df1.columns]

# 결과 출력 (데이터프레임의 첫 5행)
print(df1.head())
```

```python
# 현재 컬럼명 확인
print("현재 컬럼명 리스트:")
print(df1.columns.tolist())
```

```python
# 컬럼명 변경: 원하는 이름으로 딕셔너리 생성
new_column_names = {
    # "기존 컬럼명": "새로운 컬럼명"
    'Unnamed: 0_법정동시도(계x)': '시/도',
    'Unnamed: 1_법정동시군_일부통합': '시/군/구',
    '메트릭_Unnamed: 2_level_1': '삭제필요',
    '지급공통_월평균지급액2_Unnamed: 3_level_1': '월평균지급액'
}

# 컬럼명 변경 적용
df1.rename(columns=new_column_names, inplace=True)

# 결과 확인
print("변경된 컬럼명 리스트:")
print(df1.columns.tolist())
```

```python
df1
```

```python
# 시/도별로 그룹화하고 월평균지급액 평균 계산
grouped_df = df1.groupby('시/도')['월평균지급액'].mean().reset_index()

# 결과 확인
grouped_df
```

```python
# 파일 저장 경로 (예: '/mnt/data/전처리된_연금데이터.csv')
output_file_path = 'data/국민연금 평균 수령액_전처리.csv'

# 전처리된 데이터프레임(df_multi_header)을 CSV로 저장
grouped_df.to_csv(output_file_path, index=False, encoding='utf-8-sig')

print(f"전처리된 데이터가 {output_file_path}에 저장되었습니다!")
```
